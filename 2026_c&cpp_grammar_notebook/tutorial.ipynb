### 1. 变量本质上是什么
变量有三个重要属性：
变量名：a  //网址
变量值：10  //网站内容
变量地址：&a  //IP

### 2. &：取地址符

````c++
int a = 10;
printf("%d\n", a);   // 输出 a 的值
printf("%p\n", &a);  // 输出 a 的地址，&a 表示“取出变量 a 的地址”。
````

### 3. 指针是什么
指针就是一种变量，它存储的是地址. 一种特殊的空间，网站内容是其它IP

````c++
int a = 10;
int *p = &a;
cout << p << endl;
````

### 4. * 的两种含义

- 4.1 声明指针
    int *p; // 这里的 * 表示：p 是一个指针变量，指向 int 类型的数据。[p 是一个指针，p 里面存放某个 int 变量的地址]
- 4.2 解引用
    *p // 这里的 * 表示：访问指针 p 指向的那块内存。 // 看看这个特殊网站里的IP内容
    //因为指针变量存储内容的特殊性，不仅能窥探到存储内容（IP），这个存储内容本身也是窥探其它区域的一把钥匙，所以*p相当于开两次门，直接看到了藏在最后的幕后黑手

````c++
int a = 10;
int *p = &a;

printf("%d\n", *p);  //因为 p 存的是 a 的地址，*p 就是通过这个地址找到 a，然后读取它的值。
````

### 5. 指针的完整基本模型
看这段代码：
```cpp
int a = 10;
int *p = &a;

printf("%d\n", a);
printf("%p\n", &a);
printf("%p\n", p);
printf("%d\n", *p);
```

含义分别是：
```
a      // a 的值，10
&a     // a 的地址
p      // p 保存的地址，也就是 a 的地址
*p     // p 指向的变量的值，也就是 a 的值
```
所以：
```
p == &a
*p == a
```
如果执行：
```
*p = 20;
```
等价于：
```
a = 20;
```
因为 *p 就是 p 指向的那个变量。

完整例子：
```cpp
#include <stdio.h>

int main()
{
    int a = 10;
    int *p = &a;

    *p = 20;

    printf("%d\n", a);  // 20

    return 0;
}
```


### 6. 指针类型为什么重要

- 比如：它们都存地址，但类型不同。区别在于：它们指向的数据类型不同。
```cpp
int *p;      // p 指向 int
char *q;     // q 指向 char
double *r;   // r 指向 double
```
- 指针类型至少决定三件事：

第一，解引用时按什么类型解释内存：如果 `p` 是 `int *`，系统会把那块内存解释成 `int`。
第二，指针加减时跳过多少字节：如果 `p` 是 `int *`，`p + 1` 通常前进 4 字节。如果 `p` 是 `double *`，`p + 1` 通常前进 8 字节
第三，编译器能检查类型是否匹配：
    `int a = 10;`
    `double *p = &a;  // 类型不匹配`

### 7. 指针和数组的关系

- 数组名在很多表达式中会退化成指向首元素的指针。

例如：`int a[3] = {10, 20, 30};` 这里a在多数情况下等价于 `&a[0]`，所以 `int *p = a;` 等价于 `int *p = &a[0]; `

- 访问数组时，`a[0]` 等价于 `*(a + 0)`,同理 `a[1]` 等价于 `*(a + 1)`
